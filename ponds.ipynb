```python
# removed NA values
ponds_data = ponds_data.dropna(subset = ['Time'])
ponds_data = ponds_data.dropna(subset = ['NITRATE(PPM)'])
ponds_data = ponds_data.dropna(subset = ['PH'])
ponds_data = ponds_data.dropna(subset = ['TEMP'])
ponds_data = ponds_data.dropna(subset = ['DO'])
ponds_data = ponds_data.dropna(subset = ['MANGANESE(mg/l)'])
```

```python

# print(ponds_data.dtypes)
print(ponds_data["PH"].dtype)  # Check data type
print(ponds_data["PH"].unique())  # See unique values

ponds_data["PH"] = pd.to_numeric(ponds_data["PH"], errors="coerce")  


```

```python
# Convert the Date column to datetime format (ignoring Time)
ponds_data["Date"] = pd.to_datetime(ponds_data["Date"], format="%d-%m-%Y")

# Now you can proceed with filtering your data
start_date = ponds_data["Date"].min()
end_date = start_date + pd.Timedelta(weeks=1)  # 1-week span

relevant_data = ponds_data[
    (ponds_data["Date"] >= start_date) & (ponds_data["Date"] < end_date) &  
    (ponds_data["PH"] >= 7.5) & (ponds_data["PH"] <= 8.5) &
    (ponds_data["TEMP"] >= 26) & (ponds_data["TEMP"] <= 30) &
    (ponds_data["TURBIDITY"] >= 30) & (ponds_data["TURBIDITY"] <= 60)
]

print(relevant_data)
```
